```python
a = 3
N_hosts = 50000 # total surveyed hosts
has_udg = np.random.poisson(lam=0.4, size=(N_hosts))
n_udg = np.sum(has_udg)

a_fake = 2
contam_den = 1.5 #1.5 #1.5 # per deg2
area = 0.1 # deg2

###################################
# gen mock size data
host_ind = np.where(has_udg > 0)[0]
udg_size_dict = {}
for ind in host_ind:
    udg_size_dict[ind] = np.log10(0.3 * 1 / np.random.power(a, size=(has_udg[ind])))
    _n_contam = np.random.poisson(area * contam_den)
    if _n_contam > 0:
        _fakes = np.log10(0.3 * 1 / np.random.power(a_fake, size=(_n_contam)))
        udg_size_dict[ind] = np.concatenate([udg_size_dict[ind], _fakes])
```

```python
# generate fake udgs
fake_udg_sizes = np.log10(0.3 * 1 / np.random.power(a_fake, size=(50000)))
output, cen = moving_binned_statistic(fake_udg_sizes,
                                      fake_udg_sizes,
                                      x_err=None,
                                      bins=n_bins, 
                                      range_=[-0.5, 0.5],
                                      statistic='count', 
                                      n_slide=1)
n_cen_bkg = np.nanmean(output, axis=0) / np.diff(cen)[0] / (50000 / (contam_den))
# num of fake UDG per sqr deg per log_re
```
